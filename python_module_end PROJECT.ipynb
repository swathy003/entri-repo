### Analysis Tasks

### 1.Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

```python
#count the no of employees in each team
team_distribution = data['Team'].value_counts()
team_distribution
```

```python
#calculate the percentage split for each team
percentage_split = (team_distribution/len(data))*100
percentage_split
```

## 2.Segregate employees based on their positions within the company.

```python
employees = data.groupby('Position')['Name'].apply(list)
for Position, Names in employees.items():
    print(f"\tEmployees in Position: {Position}")
    for i in Names:
     print(i)
    print("\n")
```

## 3.Identify the predominant age group among employees.

```python
# Define age bins and labels for grouping
age_bins = [20, 30, 40, 50, 60]
age_labels = ['20-30', '31-40', '41-50', '51-60']

# Create a new column 'Age Group' based on the defined bins
data['Age Group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels)

# count the number of employees in each group
age_group_counts = data['Age Group'].value_counts()

# Find the most predominant age group
predominant_group = age_group_counts.idxmax()

 #Output the result
print("Age Group counts:\n", age_group_counts)
print("\nMost Predominant Age Group:", predominant_group)
```

### 4. Discover which team and position have the highest salary expenditure.

```python
#group by 'Team' and 'position' and sum of salaries
salary_expenditure = data.groupby(['Team','Position'])['Salary'].sum()

# Find the maximum salary expenditure and corresponding team and position
max_expenditure = salary_expenditure.idxmax()
highest_salary = salary_expenditure.max()

# Output the result
print(f"The team  and position  with the highest salary expenditure is: {max_expenditure[0]} team and {max_expenditure[1]} position with a total salary expenditure of {highest_salary}.")
```

## 5.Investigate if there's any correlation between age and salary, and represent it visually.

```python
# Calculate the correlation coefficient between Age and Salary
correlation = data['Age'].corr(data['Salary'])

# Output the correlation 
print("The correlation between Age and Salary :", correlation)
```

## Graphical Representation



## 1. Distribution of employees across each team

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Bar plot

```python
#count the no of employees in each team
team_distribution = data['Team'].value_counts()

#create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=team_distribution.index, y=team_distribution.values,color='g',width=0.5,edgecolor='black',)
plt.title('Employees Distribution by Team',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'g'})
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=90)
plt.show()
```

## 2. Distribution of employees across position

## Pie chart

```python
#count the no of employees in each position 
position_distribution = data['Position'].value_counts()

#create a bar plot
plt.figure(figsize=(10, 6))
position_distribution.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Set2', len(position_distribution)), startangle=90)
plt.title('Employees Distribution by Position',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'m'})
plt.show()
```

## 3. Age group among employees

## Histogram plot

```python
#create histogram
plt.figure(figsize=(10, 6))
plt.hist(data["Age"],bins=np.arange(20,60,10),edgecolor='black')
plt.title('Age Group among Employees',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'b'})
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

## 4. Salary expenditure by Team and Position
## Bar plot


```python
#group salary by team and position
team_position_salary = data.groupby(['Team','Position'])['Salary'].mean().reset_index()

#Create a bar chart for salary by team and position
plt.figure(figsize=(10, 6))
plt.bar(team_position_salary['Team']+'-'+team_position_salary['Position'],team_position_salary['Salary'])
plt.title(' Highest Salary by Team and Position',fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'b'})
plt.xlabel('Team-Position')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

## 5. Correlation between age and salary
## Scatter plot

```python
# Scatter plot for Age and Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Age" ,y= "Salary",data = data,color='red', edgecolor='black')
plt.ylabel("Salary")
plt.xlabel("Age")
plt.title("Correlation between Age and Salary",fontdict={'fontname':'Comic Sans MS','fontsize':20,'color':'r'})
plt.tight_layout()
plt.show()
```

## Heat map

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
data_numeric = data.select_dtypes(include=['number'])

# Drop rows with NaN values 
data_cleaned = data_numeric.dropna()

# Calculate the correlation matrix
correlation_matrix = data_cleaned.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', linewidths=0.5)
plt.title('Correlation Matrix ', fontsize=16)
plt.show()
```
